```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb
%matplotlib inline
```

```python
movie_by_years=pd.read_csv('movies_by_years.csv')
bar_movie=pd.read_csv('bar_movie.csv')
```

#### which more types produced over years ?

```python
order_type=bar_movie.variable.value_counts().index
base=sb.color_palette()[0]
sb.countplot(data=bar_movie,y='variable',color=base,order=order_type);
```

##### The chart shows that Drama , Comedy , Thriller , Action and Romance most movies in distribution of count of the movie by types

drop other types to do more analysis of them 

```python
bar_movie.drop(bar_movie.query('variable != "Drama" and variable != "Comedy" and variable != "Thriller" and variable != "Action" and variable != "Romance" ').index,inplace=True)
```

#### How many the rate of profit of movies over years ?

```python
plt.ylabel('profit')
plt.xlabel('year')
plt.title('relationship between year and profit')
plt.plot(movie_by_years['release_year'],movie_by_years['profit']);
```

##### The chart shows that sharp increas and decreas but Overall is increas over  the years.
##### and after 1990s has more increas then before

### How many the rate of popularity of movies over years ?

```python
plt.ylabel('popularity')
plt.xlabel('year')
plt.title('relationship between year and popularity')
plt.plot(movie_by_years['release_year'],movie_by_years['popularity']);
```

##### The chart shows that smooth increas over  the years conversely profit.

### what most year are produced from top five type of movies ?

```python

g = sb.violinplot(data = bar_movie, y = 'release_year',
                 x = 'variable',color=base,inner = 'quartile')
plt.title('Relationship between years and most five type movies');
```

The chart shows that the medien of most five type movies is in 2004.

### How many profit came form each type from top five of type of movies ? 

```python
pie=bar_movie.groupby('variable').sum()
pie.reset_index(inplace=True)
```

```python
plt.pie(pie.profit,labels=['Action','Comedy','Drama','Romance','Thriller'],autopct='%.1f%%');
plt.title('Relationship between profit and most five type movies');
```

##### The chart shows that Action , comedy and thriller is  most three type  have heigh profit
